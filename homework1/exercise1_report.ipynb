The Collatz Map $C(n)$ is defined to be:

$$ C(n) = \begin{cases} n / 2, & \text{if } n \equiv 0 (mod 2), \\ 
3n + 1  & \text{if } n \equiv 1 (mod 2), \\
1 & \text{if } n = 1 \end{cases} 
$$
 

Where a Collatz sequence is:

```python
import numpy as np, matplotlib.pyplot as plt
from exercise1 import collatz
%matplotlib inline
```

```python
#Create an array for the stopping times for each Collatz sequence
stoppingTime = []
for i in range(1, 5001, 1):
    seq = collatz(i)
    stoppingTime.append(len(seq) - 1)
```

```python
#Make an array of integer values 
#Plot against the corresponding stopping times
fig = plt.figure(figsize=(8, 6))
integer_n = np.arange(1, 5001, 1)
plt.scatter(integer_n, stoppingTime, c='blue')
plt.grid(b=True, which='both', color='0.65', linestyle='--')
plt.xlabel('$n$')
plt.ylabel('Stopping Time')
plt.xlim(0, 5000)
plt.ylim(0, 250)
plt.title('Scatter Plot of the Stopping Times of the Collatz Map\n plotted for the first 5000 natural numbers\n')
```
